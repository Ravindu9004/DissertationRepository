```python
import nltk
nltk.downloader.download('vader_lexicon')
from nltk.sentiment import SentimentIntensityAnalyzer
from sklearn.metrics import silhouette_score
```

```python
sia = SentimentIntensityAnalyzer()

senti_df = pd.DataFrame()


for raw in newscopy_df.heading:
    result = sia.polarity_scores(raw)
    # print(result)
    senti_df = senti_df.append(pd.DataFrame([result]))
```

```python
senti_df.head(10)
```

```python
senti_df['Polarity'] = senti_df.idxmax(axis=1)
```

```python
senti_df['Polarity confidence'] = senti_df.max(axis = 1)
senti_df
```

```python
import re
```

```python
import unicodedata
from nltk.corpus import stopwords
ADDITIONAL_STOPWORDS = ['@']

```

```python
def basic_clean(text):
  "A simple function to clean up the data."
  wnl = nltk.stem.WordNetLemmatizer()
  stopwords = nltk.corpus.stopwords.words('english')
  text = (unicodedata.normalize('NFKD', text)
    .encode('ascii', 'ignore')
    .decode('utf-8', 'ignore')
    .lower())
  words = re.sub(r'[^\w\s]', '', text).split()
  return [wnl.lemmatize(word) for word in words if word not in stopwords]
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```
